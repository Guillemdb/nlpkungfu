```python
from nltk.corpus import brown
data = []
for fileid in brown.fileids():
    document = ' '.join(brown.words(fileid))
    data.append(document)
```

```python
data = data[:3]
```

```python
from nlpkf.models.topics import TopicAnalizer
tok_kwargs = dict(remove_stopwords=True, use_stems=False, 
                  to_lowercase=True, use_lemma=True,
                  remove_punctuation=True, normalize_strings=True,
                  filter_pos=True, remove_num=True)
topics = TopicAnalizer(n_components=3, tokenizer_kwargs=tok_kwargs)
```

```python
transformed_data = topics.fit(data)
```

```python
topics.print_topics()
```

```python
from nlpkf.utils import enable_notebook_visualization

enable_notebook_visualization()
```

```python
data_vectorized = topics.vectorizer.transform(data)
```

```python
topics.visualize_topics(data_vectorized)
```

```python
topics.plot_documents(data_vectorized)
```

```python
topics.plot_words(data_vectorized)
```

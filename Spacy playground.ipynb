```python
text = u'Apple is looking at buying U.K. startup for $1 billion. My aunt elise gave two peanuts to her dog. See the thing'
doc = nlp(text)
```

```python
def explain_text_entities(text):
    doc = nlp(text)
    for ent in doc.ents:
        print(f'Entity: {ent}, Label: {ent.label_}, {spacy.explain(ent.label_)}')
    return ent
```

```python
spacy.displacy.render(doc, style='ent',jupyter=True)
```
